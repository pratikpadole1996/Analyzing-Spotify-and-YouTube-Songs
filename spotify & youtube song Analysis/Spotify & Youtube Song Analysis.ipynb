```python
# Removing Unwanted_Colmuns
def remove_unwanted_columns(df):
    unwanted_columns=['Url_spotify','Uri', 'Key', 'Url_youtube','Description']
    df.drop(columns=unwanted_columns,inplace=True)
    return df
remove_unwanted_columns(df)
```

```python
# Checking Missing_Value
df.isna().sum()
```

```python
#Treating Missing Vlaue
def treat_null(df):
    df.dropna(subset=['Danceability'],inplace=True)
    df.dropna(subset=['Energy'],inplace=True)
    df.dropna(subset=['Loudness'],inplace=True)
    df.dropna(subset=['Speechiness'],inplace=True)
    df.dropna(subset=['Acousticness'],inplace=True)
    df.dropna(subset=['Instrumentalness'],inplace=True)
    df.dropna(subset=['Liveness'],inplace=True)
    df.dropna(subset=['Valence'],inplace=True)
    df.dropna(subset=['Tempo'],inplace=True)
    df.dropna(subset=['Duration_ms'],inplace=True)
    df['Title'].fillna('unknown',inplace=True)
    df['Channel'].fillna('unknown',inplace=True)
    df['Views'].fillna(0,inplace=True)
    df['Likes'].fillna(0,inplace=True)
    df['Comments'].fillna(0,inplace=True)
    df['Licensed'].fillna('True',inplace=True)
    df['official_video'].fillna('True',inplace=True)
    df['Stream'].fillna(0,inplace=True)
treat_null(df)
```

```python
# Checking for duplicates
df.duplicated().value_counts()
```

```python
# Droping dulplicates value
df.drop_duplicates(keep='first',inplace=True)
```

```python
# Converting duration(millesecond to min) 
df['Duration_ms']=df['Duration_ms']/60000
```

```python
# As we convert ms to min our column name should also be chnage from Duration_ms to Duration_min
df.rename(columns={'Duration_ms': 'Duration_min'}, inplace=True)
```
